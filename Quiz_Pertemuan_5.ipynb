```python
# Soal no.5

print("Daftar bilangan genap dan hasil kuadratnya:")
for i in range(1, 51):
    if i % 2 == 0:
        print(f"{i}^2 = {i**2}")

print("\n")  # baris kosong untuk pemisah

i = 1
total = 0
while i <= 50:
    if i % 2 == 0:
        total += i**2
    i += 1

print("Jumlah seluruh kuadrat bilangan genap dari 1 sampai 50 adalah:", total)
```
