```python
# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

# Contoh pemanggilan fungsi (bonus)
hasil = hitung_efisiensi(500, 400)
print("Efisiensi turbin =", hasil)
```

```python
# Fungsi/Modularisasi
def hitung_efisiensi_termal():
    energi_masuk = float(input("Masukkan energi masuk (Joule): "))
    energi_keluar = float(input("Masukkan energi keluar (Joule): "))

    efisiensi = (energi_keluar / energi_masuk) * 100

    print(f"Efisiensi termal turbin gas = {efisiensi:.2f}%")

hitung_efisiensi_termal()
```

```python
# Challenge

# === PROGRAM EFISIENSI ENERGI LAMPU ===

def hitung_energi(watt, jam_per_hari, hari):
    return watt * jam_per_hari * hari / 1000  # hasil dalam kWh

def main():
    print("=== PROGRAM EFISIENSI ENERGI LAMPU ===")
    total_kwh = 0
    total_biaya = 0
    tarif = 1500  # Rp per kWh

    jumlah_lampu = int(input("Masukkan jumlah lampu di rumah: "))

    for i in range(1, jumlah_lampu + 1):
        print(f"\nLampu ke-{i}")
        watt = float(input("Daya lampu (Watt): "))
        jam = float(input("Rata-rata jam nyala per hari: "))

        energi = hitung_energi(watt, jam, 30)
        biaya = energi * tarif
        total_kwh += energi
        total_biaya += biaya

        if watt <= 10:
            kategori = "Hemat Energi üí°"
        elif watt <= 20:
            kategori = "Sedang ‚ö°"
        else:
            kategori = "Boros ‚ö†Ô∏è"

        print(f"Kategori: {kategori}")
        print(f"Pemakaian listrik per bulan: {energi:.2f} kWh (Rp {biaya:.0f})")

    print("\n=== RINGKASAN ===")
    print(f"Total konsumsi: {total_kwh:.2f} kWh")
    print(f"Total biaya listrik: Rp {total_biaya:.0f}")

    while True:
        saran = input("\nIngin saran penghematan? (ya/tidak): ").lower()
        if saran == "ya":
            if total_kwh > 50:
                print("Gunakan lampu LED dan matikan lampu saat tidak digunakan üí°")
            else:
                print("Penggunaan lampu sudah efisien üëç")
            break
        elif saran == "tidak":
            print("Baik, terima kasih!")
            break
        else:
            print("Masukkan hanya 'ya' atau 'tidak'.")

main()
```
